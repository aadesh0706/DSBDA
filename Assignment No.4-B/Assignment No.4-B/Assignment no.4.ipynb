```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
adult = pd.read_csv('adult.csv') 
```

```python
adult.head()
```

```python
adult.info()
```

```python
adult.isnull().sum()
```

```python
print(adult.isnull().sum())
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
categorical_cols = adult.select_dtypes(include='object').columns

for col in categorical_cols:
    adult[col] = le.fit_transform(adult[col])
```

```python
print(adult.head())
print(adult.isnull().sum())
```

```python
plt.figure(figsize=(8,5))
plt.hist(adult['age'], bins=30, color='lightgreen', edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='income', y='hours.per.week', data=adult, palette='coolwarm')
plt.title("Hours per Week by Income")
plt.xlabel("Income Group")
plt.ylabel("Hours per Week")
plt.show()
```

```python
# Select a few features to visualize
features = ['age', 'education.num', 'hours.per.week', 'income']
sns.pairplot(adult[features], hue='income', palette='Pastel2')
plt.suptitle("Pairplot of Age, Education, Hours, Income", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='income', data=adult, palette=['#f7b7a3', '#a1cfff'])
plt.title("Income Group Distribution", fontsize=14)
plt.xlabel("Income Group")
plt.ylabel("Count")
plt.show()
```
